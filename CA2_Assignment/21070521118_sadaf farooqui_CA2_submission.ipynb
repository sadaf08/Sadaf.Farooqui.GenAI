Q:2 Generate a model in Python to represent a Housing loan scheme and create a chart to
display the Emi based on rate of interest and reducing balance for a given period. If a customer
wishes to close the loan earlier, print the interest lost distributed over the remaining no. Of
months. Assume suitable data and inputs as necessary.

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
class HousingLoan:
    def __init__(self, principal, annual_rate, tenure_years):
        self.principal = principal
        self.annual_rate = annual_rate
        self.tenure_years = tenure_years
        self.monthly_rate = annual_rate / 12 / 100
        self.tenure_months = tenure_years * 12

    def calculate_emi(self):
        """Calculate EMI using the formula"""
        P = self.principal
        r = self.monthly_rate
        n = self.tenure_months
        emi = (P * r * (1 + r)**n) / ((1 + r)**n - 1)
        return emi

    def get_reducing_balance(self):
        """Calculate the reducing balance over time"""
        emi = self.calculate_emi()
        balance = self.principal
        balances = []

        for _ in range(self.tenure_months):
            interest_payment = balance * self.monthly_rate
            principal_payment = emi - interest_payment
            balance -= principal_payment
            balances.append(balance)

        return balances
```

```
principal = float(input("Enter the principal amount of the loan: "))
annual_rate = float(input("Enter the annual interest rate (as a percentage): "))
tenure_years = int(input("Enter the loan tenure (in years): "))

loan = HousingLoan(principal, annual_rate, tenure_years)

emi = loan.calculate_emi()
reducing_balance = loan.get_reducing_balance()
print(f"EMI: {emi:.2f}")
```

```
interest_rates = np.arange(3, 11, 0.5)
emis = []

for rate in interest_rates:
    temp_loan = HousingLoan(principal, rate, tenure_years)
    emi = temp_loan.calculate_emi()
    emis.append(emi)
```

```
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(interest_rates, emis, marker='o', color='green')
plt.title('Emi based on rate of interest')
plt.xlabel('Annual Interest Rate (%)')
plt.ylabel('EMI Amount')
plt.grid(True)

months = np.arange(1, len(reducing_balance) + 1)
plt.subplot(1, 2, 2)
plt.plot(months, reducing_balance, label='Reducing Balance', color='blue')
plt.title('Reducing Balance Over Time')
plt.xlabel('Month')
plt.ylabel('Remaining Balance')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

Q:5 Generate a model for Covid 19 with symptoms of parameters like fever, cold, shivering, weight loss, generate 100 model data with random values for each parameter and order by parameter lowest to highest in display based on the input parameter.

```
import numpy as np
import pandas as pd
```

```
parameters = ['fever', 'cold', 'shivering', 'weight_loss']
num_samples = 100
```

```
data = np.random.randint(0, 11, size=(num_samples, len(parameters)))
df = pd.DataFrame(data, columns=parameters)
```

```
def sort_by_parameter(df, parameter):
    if parameter in parameters:
        return df.sort_values(by=parameter)
    else:
        return f"Invalid parameter. Choose from {parameters}"
```

```
pd.set_option('display.max_rows', None)
parameter_to_sort = input(f"Sort by one of {parameters}: ")
sorted_df = sort_by_parameter(df, parameter_to_sort)
print(sorted_df)
```
